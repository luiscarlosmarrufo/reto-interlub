## 1. Formulación matricial de RLM

Nuestro modelo es

$$y_i=\beta_0+\beta_1x_{i,1}+...+\beta_kx_{i,k}+\epsilon_i$$

```python
# aqui se empieza el 
np.random.seed(2025) 

data = pd.read_csv("../../datos/expanded_data.csv")
data = data.replace(-99, np.nan)

n = len(data)
k = 4
p = k + 1
```

```python
num = data.select_dtypes(include=[np.number]).copy()
# quitar columnas "color", "Unnamed: 0", " textura", "Aceite Base", "Registro NSF", "Espesante"
num = num.drop(columns=["color", "Unnamed: 0", "textura", "Aceite Base", "Registro NSF", "Espesante"], axis=1)
num.info()
```

```python
corr_abs = num.corr().abs()

avg_abs_corr = corr_abs.mean().sort_values(ascending=False)
print("\nTop variables por correlación media absoluta:")
display(avg_abs_corr.head(20))
```

```python
rlm_vars = ['Viscosidad del Aceite Base a 40°C. cSt', 'Penetración de Cono a 25°C, 0.1mm', 'Estabilidad Mecánica, %', 'Punto de Gota, °C', 'Carga Timken Ok, lb']
```

```python
X1 = num[rlm_vars[0]]
X2 = num[rlm_vars[1]]
X3 = num[rlm_vars[2]]
X4 = num[rlm_vars[3]]
X5 = num[rlm_vars[4]]
```

```python
X = np.column_stack([np.ones(n), X1, X2, X3, X4, X5])
```

```python
#col_mean = np.nanmean(X, axis=0)
#indices = np.where(np.isnan(X))
#X[indices] = np.take(col_mean, indices[1])

mask = ~np.isnan(X).any(axis=1)

# aplicar la máscara
X = X[mask]
```

```python
n = len(X)
```

```python
for i in range(p):
    y = X[:,i+1]
    X_i = np.delete(X, i+1, axis=1)

    XtX = X_i.T @ X_i
    XtX_inv = np.linalg.inv(XtX)
    XtY = X_i.T @ y

    beta_hat = XtX_inv @ XtY
    print(beta_hat)

    y_hat = X_i @ beta_hat
    res = y - y_hat
    #print(pd.DataFrame(res).describe())

    SSE = float(res.T @ res)
    SSR = float(((y_hat - y.mean())**2).sum())
    TSS = float(((y - y.mean())**2).sum())

    MSE = SSE / (n - p)
    MSR = SSR / (p - 1)
    F_stat = MSR / MSE
    R2 = 1 - SSE / TSS
    R2_adj = 1 - (1-R2)*(n-1)/(n-p)

    summary = {
    "SSE": SSE,
    "SSR": SSR,
    "TSS": TSS,
    "MSE": MSE,
    "MSR": MSR,
    "F_stat": F_stat,
    "R2": R2,
    "R2_adj": R2_adj
    }

    display(pd.DataFrame(summary, index=[0]).T.style.set_caption("Resumen del modelo"))
```

```python
y = X[:,5]
X_final = np.delete(X, 5, axis=1)

XtX = X_final.T @ X_final
XtX_inv = np.linalg.inv(XtX)
XtY = X_final.T @ y

beta_hat = XtX_inv @ XtY

y_hat = X_final @ beta_hat
res = y - y_hat

MSE = np.mean(res**2)
```

```python
# construya la matriz X^T X
X_transpose_X = X_final.T @ X_final
print(f"\nX'X matrix ({p} × {p}):")
print(X_transpose_X)

# primero verificar si X^T X es invertible con el rango
rank = np.linalg.matrix_rank(X_transpose_X)
print(f"\nRango of X'X: {rank}/{p}")
print(f"X'X es invertible: {rank == p}")
```

## 4. Estimador de $y (\hat{y})$

Ejercicios:
- Estime y con las lineas de abajo
- Analice y valide en el contexto de grasas si sus resultados van acorde a lo esperado


```python
# ahora encuentre la inversa de X^T X
X_transpose_X_inv = np.linalg.inv(X_transpose_X)

# encuentra beta vector estimador
beta_hat = X_transpose_X_inv @ X_final.T @ y
#print(f"True β: {true_beta}")
print(f"OLS β_hat: {beta_hat}")

# encuentre la matriz de covarianza de los estimadores
cov_beta_hat = (sigma ** 2) * X_transpose_X_inv
print(f"\nTheoretical Cov(β_hat):")
print(cov_beta_hat)

# encuentre los errores estandar de las b0, b1,b2... extienda hasta donde usted decido anteriormente
std_errors = np.sqrt(np.diag(cov_beta_hat))
print(f"\nTheoretical Standard Errors:")
print(f"SE(β₀): {std_errors[0]:.4f}")
print(f"SE(β₁): {std_errors[1]:.4f}")
print(f"SE(β₂): {std_errors[2]:.4f}")

# encuentre el estimador de y
y_hat = X_final@X_transpose_X_inv @ X_final.T @ y
```

## 7. F-estadistica de RLM

Ejercicios:
- Complete las siguientes "leads" de código
- Adapte el codigo para dar solucion al reto Interlub

```python
# aqui se empieza el código

# dada la matriz X de la seccion anterior
# muestre que I - X(X'X)-1 X' es idempotente
I = np.eye(n)
I_idem = I - (X_final @ XtX_inv @ X_final.T)

result = np.allclose(I_idem @ I_idem, I_idem)
print(result)

# encuentre el rango I - X(X'X)-1 X'
# [Sugerencia: rango = np.linalg.matrix_rank()]
rango = np.linalg.matrix_rank(I_idem)
print(rango)

# muestre que X(X'X)-1 X' es idempotente
X_idem = X_final @ XtX_inv @ X_final.T

result = np.allclose(X_idem @ X_idem, X_idem)
print(result)

# encuentre el rango de X(X'X)-1 X'
# [Sugerencia: rango = np.linalg.matrix_rank()]
rango = np.linalg.matrix_rank(X_idem)
print(rango)


# muestre que (I - X(X'X)-1 X')*(X(X'X)-1 X) = 0
# cuales son las consecuencias de esto
result = np.allclose(I_idem @ X_idem, 0)
print(result)

# ahora usando la y vector de la seccion anterior
# y las betas y sus estimadores
# encuentre la estadística lambda (resultado del Lemma de Neyman-Pearson)
```
