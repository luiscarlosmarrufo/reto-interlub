## Mejoras

```python
# Crear un soup mejorado combinando todas las descripciones
descripcion_grasas['soup'] = (
    descripcion_grasas['subtitulo'].fillna('') + ' ' +
    descripcion_grasas['descripcion'].fillna('') + ' ' +
    descripcion_grasas['beneficios'].fillna('') + ' ' +
    descripcion_grasas['aplicaciones'].fillna('')
)

descripcion_grasas['soup'] = descripcion_grasas['soup'].str.strip()
descripcion_grasas.head()
```

```python
# TF-IDF sobre el soup completo
tf_improved = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), 
                              min_df=1, stop_words=stop_words_espanol)
tfidf_matrix_improved = tf_improved.fit_transform(descripcion_grasas['soup'])

print(f"Forma de la matriz TF-IDF mejorada: {tfidf_matrix_improved.shape}")
```

```python
# Calcular similitud del coseno con el soup mejorado
cosine_sim_improved = linear_kernel(tfidf_matrix_improved, tfidf_matrix_improved)
print(f"Matriz de similitud: {cosine_sim_improved.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

# Seleccionar solo características numéricas relevantes
numeric_cols = df_v.select_dtypes(include='number').columns.tolist()
print(f"Columnas numéricas: {numeric_cols}")

# Crear DataFrame con características numéricas normalizadas
df_numeric = df_v[numeric_cols].copy()
df_numeric = df_numeric.fillna(df_numeric.median())  # Rellenar NaN con mediana

# Normalizar
scaler = StandardScaler()
numeric_scaled = scaler.fit_transform(df_numeric)

# Calcular similitud basada en características numéricas
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim_numeric = cosine_similarity(numeric_scaled)

print(f"Matriz de similitud numérica: {cosine_sim_numeric.shape}")
```

```python
# Combinar ambas similitudes con pesos
# 70% texto + 30% características numéricas
peso_texto = 0.7
peso_numerico = 0.3

cosine_sim_hybrid = (peso_texto * cosine_sim_improved) + (peso_numerico * cosine_sim_numeric)

print(f"Matriz de similitud híbrida: {cosine_sim_hybrid.shape}")
print(f"Ejemplo de scores híbridos para primera grasa: {cosine_sim_hybrid[0][:5]}")
```

```python
def get_recommendations_improved(codigo_grasa, top_n=10, use_hybrid=True):
    """
    Obtiene recomendaciones basadas en similitud de contenido mejorada
    
    Parameters:
    -----------
    codigo_grasa : str
        Código de la grasa de referencia
    top_n : int
        Número de recomendaciones a retornar
    use_hybrid : bool
        Si True, usa similitud híbrida (texto + numérico)
        Si False, usa solo similitud de texto
    """
    # Seleccionar matriz de similitud
    sim_matrix = cosine_sim_hybrid if use_hybrid else cosine_sim_improved
    
    # Obtener índice de la grasa
    idx = indices[codigo_grasa]
    
    # Obtener scores de similitud
    sim_scores = list(enumerate(sim_matrix[idx]))
    
    # Ordenar por similitud (descendente)
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Excluir la grasa misma (índice 0) y tomar top_n+1
    sim_scores = sim_scores[1:top_n+1]
    
    # Obtener índices de las grasas recomendadas
    grasa_indices = [i[0] for i in sim_scores]
    
    # Crear DataFrame con resultados
    recommendations = df_v_id.iloc[grasa_indices][['codigoGrasa']].copy()
    recommendations['similarity_score'] = [score[1] for score in sim_scores]
    
    return recommendations
```

```python
# Probar con una grasa específica
print("=== Recomendaciones solo por TEXTO ===")
recs_text = get_recommendations_improved("Grasa_10", top_n=5, use_hybrid=False)
print(recs_text)

print("\n=== Recomendaciones HÍBRIDAS (texto + propiedades) ===")
recs_hybrid = get_recommendations_improved("Grasa_10", top_n=5, use_hybrid=True)
print(recs_hybrid)
```

```python
def calculate_weighted_score(df_input):
    """
    Calcula un score ponderado considerando múltiples propiedades técnicas
    Cuanto mayor el score, mejor la grasa (en general)
    """
    df_score = df_input.copy()
    
    # Normalizar cada métrica entre 0 y 1
    # Para métricas donde "mayor es mejor"
    if 'Penetracion Trabajada' in df_score.columns:
        penetration = df_score['Penetracion Trabajada'].fillna(df_score['Penetracion Trabajada'].median())
        df_score['score_penetration'] = (penetration - penetration.min()) / (penetration.max() - penetration.min())
    else:
        df_score['score_penetration'] = 0.5
    
    # Para métricas donde "menor es mejor" (invertir)
    if 'Punto de Goteo, C' in df_score.columns:
        dropping_point = df_score['Punto de Goteo, C'].fillna(df_score['Punto de Goteo, C'].median())
        df_score['score_dropping'] = 1 - ((dropping_point - dropping_point.min()) / (dropping_point.max() - dropping_point.min()))
    else:
        df_score['score_dropping'] = 0.5
    
    # Registro NSF es binario (ya está normalizado)
    if 'Registro NSF' in df_score.columns:
        df_score['score_nsf'] = df_score['Registro NSF']
    else:
        df_score['score_nsf'] = 0
    
    # Score final ponderado (ajusta pesos según importancia)
    df_score['weighted_score'] = (
        0.4 * df_score['score_penetration'] +
        0.4 * df_score['score_dropping'] +
        0.2 * df_score['score_nsf']
    )
    
    return df_score

# Aplicar a todo el dataset
df_v_scored = calculate_weighted_score(df_v_id)
df_v_scored[['codigoGrasa', 'weighted_score', 'score_penetration', 'score_dropping', 'score_nsf']].head(10)
```

```python
def hybrid_recommendations(codigo_grasa, top_n=10, use_weighted=True):
    """
    Sistema de recomendación híbrido que combina:
    1. Similitud de contenido (texto + numérico)
    2. Score ponderado de calidad técnica
    
    Parameters:
    -----------
    codigo_grasa : str
        Código de la grasa de referencia
    top_n : int
        Número de recomendaciones a retornar
    use_weighted : bool
        Si True, re-rankea por weighted_score
    """
    # 1. Obtener candidatos por similitud (más candidatos de los necesarios)
    idx = indices[codigo_grasa]
    sim_scores = list(enumerate(cosine_sim_hybrid[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Tomar top 25 candidatos
    sim_scores = sim_scores[1:26]
    grasa_indices = [i[0] for i in sim_scores]
    
    # 2. Crear DataFrame con candidatos
    candidates = df_v_scored.iloc[grasa_indices].copy()
    candidates['similarity_score'] = [score[1] for score in sim_scores]
    
    # 3. Si use_weighted, re-rankear por weighted_score
    if use_weighted:
        # Combinar similarity y weighted_score
        # Normalizar similarity_score para que esté en [0,1]
        candidates['similarity_normalized'] = candidates['similarity_score']
        
        # Score final: 60% similitud + 40% calidad técnica
        candidates['final_score'] = (
            0.6 * candidates['similarity_normalized'] +
            0.4 * candidates['weighted_score']
        )
        
        candidates = candidates.sort_values('final_score', ascending=False)
    else:
        # Solo ordenar por similitud
        candidates = candidates.sort_values('similarity_score', ascending=False)
    
    # Seleccionar columnas relevantes para mostrar
    result_cols = ['codigoGrasa', 'similarity_score', 'weighted_score', 'final_score', 
                   'Punto de Goteo, C', 'Penetracion Trabajada', 'Registro NSF']
    
    # Filtrar solo las columnas que existen
    result_cols = [col for col in result_cols if col in candidates.columns]
    
    return candidates[result_cols].head(top_n)
```

```python
# Probar con diferentes grasas
print("="*70)
print("RECOMENDACIONES PARA: Grasa_10")
print("="*70)

print("\n--- Solo por Similitud ---")
recs_similarity = hybrid_recommendations("Grasa_10", top_n=5, use_weighted=False)
print(recs_similarity)

print("\n--- Híbrido (Similitud + Score Técnico) ---")
recs_hybrid = hybrid_recommendations("Grasa_10", top_n=5, use_weighted=True)
print(recs_hybrid)
```

```python
def filter_recommendations(codigo_grasa, top_n=10, 
                          max_dropping_point=None,
                          min_penetration=None,
                          require_nsf=None):
    """
    Obtiene recomendaciones con filtros técnicos aplicados
    
    Parameters:
    -----------
    codigo_grasa : str
        Código de la grasa de referencia
    top_n : int
        Número de recomendaciones finales
    max_dropping_point : float, optional
        Punto de goteo máximo permitido
    min_penetration : float, optional
        Penetración trabajada mínima
    require_nsf : bool, optional
        Si True, solo grasas con registro NSF
    """
    # Obtener más candidatos de los necesarios (50)
    candidates = hybrid_recommendations(codigo_grasa, top_n=50, use_weighted=True)
    
    # Aplicar filtros
    if max_dropping_point is not None and 'Punto de Goteo, C' in candidates.columns:
        candidates = candidates[candidates['Punto de Goteo, C'] <= max_dropping_point]
    
    if min_penetration is not None and 'Penetracion Trabajada' in candidates.columns:
        candidates = candidates[candidates['Penetracion Trabajada'] >= min_penetration]
    
    if require_nsf is not None and 'Registro NSF' in candidates.columns:
        candidates = candidates[candidates['Registro NSF'] == (1 if require_nsf else 0)]
    
    return candidates.head(top_n)
```

```python
print("="*70)
print("RECOMENDACIONES CON FILTROS")
print("="*70)

# Sin filtros
print("\n--- Sin Filtros ---")
recs_no_filter = filter_recommendations("Grasa_10", top_n=5)
print(recs_no_filter)

# Con filtro de NSF
print("\n--- Solo con Registro NSF ---")
recs_nsf = filter_recommendations("Grasa_10", top_n=5, require_nsf=True)
print(recs_nsf)

# Con filtros múltiples
print("\n--- Punto Goteo < 200 y Penetración > 250 ---")
recs_multi = filter_recommendations("Grasa_10", top_n=5, 
                                   max_dropping_point=200,
                                   min_penetration=250)
print(recs_multi)
```

```python
def compare_grasas(codigo_original, codigos_recomendadas):
    """
    Compara la grasa original con las recomendadas
    """
    # Obtener datos
    original = df_v_scored[df_v_scored['codigoGrasa'] == codigo_original].iloc[0]
    recomendadas = df_v_scored[df_v_scored['codigoGrasa'].isin(codigos_recomendadas)]
    
    # Propiedades a comparar
    props = ['Punto de Goteo, C', 'Penetracion Trabajada', 'weighted_score']
    props = [p for p in props if p in df_v_scored.columns]
    
    # Crear DataFrame de comparación
    comparison = pd.DataFrame()
    comparison['Grasa'] = ['ORIGINAL'] + codigos_recomendadas.tolist()
    
    for prop in props:
        values = [original[prop]] + recomendadas[prop].tolist()
        comparison[prop] = values
    
    return comparison

# Probar comparación
recs = hybrid_recommendations("Grasa_10", top_n=3, use_weighted=True)
codigos_recs = recs['codigoGrasa'].tolist()

comparison_df = compare_grasas("Grasa_10", codigos_recs)
print("\n=== COMPARACIÓN ===")
print(comparison_df)
```

```python
def plot_comparison(codigo_original, top_n=5):
    """
    Grafica comparación entre grasa original y recomendaciones
    """
    # Obtener recomendaciones
    recs = hybrid_recommendations(codigo_original, top_n=top_n, use_weighted=True)
    codigos_recs = recs['codigoGrasa'].tolist()
    
    # Comparar
    comparison = compare_grasas(codigo_original, codigos_recs)
    
    # Seleccionar propiedades numéricas para graficar
    numeric_props = comparison.select_dtypes(include='number').columns.tolist()
    
    if len(numeric_props) == 0:
        print("No hay propiedades numéricas para graficar")
        return
    
    # Crear subplots
    n_props = len(numeric_props)
    fig, axes = plt.subplots(1, n_props, figsize=(5*n_props, 5))
    
    if n_props == 1:
        axes = [axes]
    
    for i, prop in enumerate(numeric_props):
        comparison.plot(x='Grasa', y=prop, kind='bar', ax=axes[i], legend=False)
        axes[i].set_title(prop)
        axes[i].set_xlabel('')
        axes[i].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

# Graficar
plot_comparison("Grasa_10", top_n=5)
```
